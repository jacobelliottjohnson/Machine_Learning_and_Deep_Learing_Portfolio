```python
import pandas as pd
import requests
import json
import os
```

```python
from elasticsearch import Elasticsearch
```

```python
directory = 'c:/Users\jacob\Documents\Data\metadata.csv\document_parses\\thousand_pdf_json/'
res = requests.get('http://localhost:9200')
print (res.content)
elastic = Elasticsearch([{'host': 'localhost', 'port': '9200'}])
```

```python
i = 0
json_files = []
for filename in os.listdir(directory):
        f = open(directory + filename)
        json_files.append(filename)
        content = f.read()
        elastic.index(index='covid', ignore=400, id=i, body=json.loads(content))
        i = i + 1
```

```python
df_metadata = pd.read_csv("c:/Users\jacob\Documents\Data\metadata.csv\metadata.csv")
df_metadata
```

```python
nes_json = []
docu = 'document_parses/pdf_json/'
for file in json_files:
    fin_doc = docu + file 
    nes_json.append(fin_doc)
df_files = pd.DataFrame(nes_json)
df_files.rename(columns = {0:'pdf_json_files'}, inplace = True)
df_files
```

```python

df_meta_1000 = pd.merge(df_metadata, df_files, on = 'pdf_json_files')
df_meta_1000
```
